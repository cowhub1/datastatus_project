### 신재생에너지_발전량과_발전비율

```python
file_path = "./data/신재생에너지_발전량과_발전비율_excel.xlsx"
energy_power = pd.read_excel(file_path,
              header=2,
             skipfooter=5,
             usecols="A:K")
energy_power
```

```python
energy_power.columns.values[0] = "연도별"
energy_power
```

```python
energy_power_t=energy_power.transpose()
energy_power_t
```

```python
energy_power_t2 = energy_power_t.rename(columns=energy_power_t.iloc[0])
energy_power_t2
```

```python
energy_power_t2 = energy_power_t2.iloc[1:,:]
```

```python
energy_power_t2
```

```python
energy_power_year=energy_power_t.index[:]
energy_power_year
```

```python
energy_power_t["연도별"] = energy_power_year
energy_power_t
```

```python
energy_power_t=energy_power_t.reset_index(drop=True)
energy_power_t
```

```python
co2_first=pd.read_excel("./data/업종별_에너지원별_CO₂배출량_현황_2004_2007.xlsx")
co2_first
```

```python
co2=pd.read_excel("./data/업종별_에너지원별_CO₂배출량_현황_2010_2011.xlsx")
```

```python
co2
```

```python
### 컬럼명 통일시키기
co2.columns = [col.split('.')[0] for col in co2.columns]
```

```python
co2
```

```python
### co2 데이터프레임에서 중복된 열을 제거하고, 첫 번째 열의 데이터만 남기는 작업
co2_sum=co2.loc[:, ~co2.columns.duplicated(keep='first')]
co2_sum
```

```python
co2_sum2=co2_sum[co2_sum["에너지원별(2)"]=="소계"].reset_index(drop=True)
co2_sum2
```

```python
co2_second=co2_sum2.iloc[[1,2,3,6]].reset_index(drop=True)
co2_second
```

```python
co2_second.iloc[:,2:]
```

```python
co2_first.iloc[:4,:]
```

```python
### co2_first와 co2_second 합치기(concat)
co2_tot=pd.concat([co2_first.iloc[:4,:], co2_second.iloc[:,2:]], axis=1)
```

```python
co2_tot=co2_tot.iloc[:,1:]
co2_tot
```

```python
###결측치 확인

co2_tot2=co2_tot.iloc[:,1:].astype(float)
```

```python
co2_tot2.info()
```

```python
co2_tot3=pd.concat([co2_tot.iloc[:,0],co2_tot2],axis=1)
```

```python
co2_tot3
```

```python
co2_tp=co2_tot3.transpose()
co2_tp
```

```python
co2_tp.rename(columns=co2_tp.loc["에너지원별",:], inplace=True)
```

```python
co2_line=co2_tp.drop("에너지원별")
co2_line
```

```python
co2_line.drop('가스류합', axis=1, inplace=True)
```

```python
co2_line
```

```python
fig = px.line(co2_line, title='에너지원별 CO2 배출량')
fig.update_layout(xaxis_title='연도', yaxis_title='배출량(천TCO2)')
fig.show()
```

```python
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import font_manager, rc

plt.rc("font", family = "Malgun Gothic")

plt.rcParams["axes.unicode_minus"] = False
```

```python
### 전국 대기오염 물질
air_pollutants = pd.read_csv("./data/전국_대기오염물질_배출량.csv")
air_pollutants
```

```python
air_pollutants=air_pollutants.transpose()
air_pollutants
```

```python
air_pollutants.rename(columns={'구분(1)': '구분'}, inplace=True)
air_pollutants
```

```python

# 데이터 타입을 숫자로 변환 (문자열이 포함되어 있으면 숫자로 변환 가능하도록 처리)
df = air_pollutants.apply(pd.to_numeric, errors='coerce')

# 데이터프레임을 전치하여 행과 열을 교환
df_transposed = df.T

# 선 그래프 그리기
fig=plt.figure(figsize=(100, 30))
df_transposed.plot(marker='o')


plt.title('전국 대기오염물질 배출량')

### x축 및 y축 제목 넣기
plt.xlabel('연도별')
plt.ylabel('대기오염 농도')
# plt.legend(title='')


### x축의 값의 기울기를 이용하여 조정하기
# xticks() : x축을 컨트롤하는 함수
# yticks() : y축을 컨트롤하는 함수
plt.xticks(rotation=30)

### 격자선 표시하기
plt.grid(True)
plt.legend(loc='upper right')
plt.savefig("./img/plant.png")

plt.show()
```
